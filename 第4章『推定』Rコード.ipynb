## p.17

```r
time <- c(1, 2, 4, 5)
status <- c(1, 1, 0, 1)
data <- cbind(time, status)
data <- data.frame(data)
```

```r
library(survival)
fit <- survfit(Surv(time, status)~1, data=data)
plot(fit)
summary(fit)
```
